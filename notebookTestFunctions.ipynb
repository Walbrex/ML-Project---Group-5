### First case : Banknote Authentification Dataset 

```python
# First we import all the functions we are going to use, from the .py file

from functionsFile import importData, cleanData, splitData, trainKnn, trainRfc, testModel 
```

```python
# First case : Banknote

file = 'data_banknote_authentication.txt'
```

```python
# We import the dataset and its main caracteristics

banknotes, banknotesShape, banknotesHead = importData(file)
```

```python
banknotesShape
```

```python
banknotesHead
```

```python
# We clean the data (remove dirtiness, centralize, normalize, ...)

banknotes, y = cleanData(banknotes)
```

```python
banknotes
```

```python
# We split the dataset into 2 parts : a train set & a test set

X_train,X_test,y_train,y_test = splitData(banknotes,y)
```

```python
# We train a KNN model

knn_gscv, best_params, best_score = trainKnn(X_train,y_train)
```

```python
# We test the model and display the main performance indicators

testModel(knn_gscv, X_test, y_test)
```

```python
# We train a Random Forest Classifier model

rfc, X_test = trainRfc(X_train,y_train,featureselection=True,t=0.15,X_test=X_test)
```

```python
# We test the model and display the main performance indicators

testModel(rfc, X_test, y_test)
```
